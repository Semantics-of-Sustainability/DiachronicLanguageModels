```python
import pandas as pd
import plotly.express as px
```

# Visualization with scatterplot
Visualize the outcomes of a contextualized embedding with an interactive scatterplot. This embedding was created in the notebook `ContextualizedEmbeddingsMacBERTH.ipynb`.

```python
datapath = '../../sentences_out/fair_embedded_macberth.csv'
df = pd.read_csv(datapath)
```

```python
fig = px.scatter(df, x='tsne1', y='tsne2', color='year', 
          hover_data=['word_{}'.format(i+1) for i in range(5)])
fig.show()
```
